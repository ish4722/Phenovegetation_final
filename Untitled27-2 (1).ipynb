```python
from google.colab import drive
import tensorflow as tf
import json
from skimage.draw import polygon
from skimage import io
import os
import glob
from tqdm import tqdm
import numpy as np
import albumentations as A
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

drive.mount('/content/drive')

def fill_mask(blobs, mask, label):
    for l in blobs:
        fill_row_coords, fill_col_coords = polygon(l[1], l[0], l[2])
        mask[fill_row_coords, fill_col_coords] = label

def convert_annotation_json_to_mask(path_to_annotation_json, path_to_masks_folder, classes, label_value, pixel_value):
    with open(path_to_annotation_json) as f:
        objects = json.load(f)
    annotations = objects['shapes']
    h, w = objects['imageHeight'], objects['imageWidth']
    mask = np.zeros((h, w)).astype('uint8')

    for annot in annotations:
        label = annot['label']
        points = annot['points']

        x_coord, y_coord, l = [], [], []
        for p in points:
            x_coord.append(int(p[0]))
            y_coord.append(int(p[1]))

        l.append(np.array(x_coord))
        l.append(np.array(y_coord))
        l.append((h, w))

        if label not in label_value:
            label_value[label] = pixel_value[0]
            pixel_value[0] += 100

        if label not in classes:
            classes[label] = [l]
        else:
            classes[label].append(l)

    for label in classes:
        fill_mask(classes[label], mask, label_value[label])
        classes[label] = []

    io.imsave(path_to_masks_folder, mask)

def get_classes(dataset_path):
    classes, label_value, pixel_value = {}, {}, [100]
    count = 0

    for dirpath, _, _ in os.walk(dataset_path):
        path_to_annotation_json = glob.glob(dirpath + "/*.json")
        for json_file in path_to_annotation_json:
            path_to_mask_png = json_file[:-5] + "_mask.png"
            convert_annotation_json_to_mask(json_file, path_to_mask_png, classes, label_value, pixel_value)
            count += 1

    assert count, "Dataset folder path does not contain any JSON mask file."
    print(".png file of all JSON mask files saved in respective folders!")

    labels = [1] * len(label_value)
    for label, value in label_value.items():
        labels[(value - 100) // 100] = label

    class_list = "$".join(labels)
    return class_list

classes = get_classes(r'/content/drive/MyDrive/test_vegetation_model')
print('Classes names:', classes.split('$'))
```

```python
import cv2
class Dataset(tf.keras.utils.Sequence):
    def __init__(self, dataset_path, n_classes, augmentation=None):
        self.images_fps, self.mask_fps = [], []
        self.n_classes = n_classes

        for dirpath, _, _ in os.walk(dataset_path):
            mask_src = glob.glob(dirpath + '/*.png')
            img_src = glob.glob(dirpath + '/*.JPG') + glob.glob(dirpath + '/*.jpeg')

            mask_src.sort()
            img_src.sort()

            if len(img_src):
                assert len(mask_src) != 0, f"{dirpath} does not contain any annotation mask file for the images."
                if len(mask_src) != 1:
                    assert len(mask_src) == len(img_src), (
                        f"{dirpath} contains {len(mask_src)} mask files but {len(img_src)} images."
                    )
                    self.mask_fps += mask_src
                else:
                    self.mask_fps += [mask_src[0]] * len(img_src)
                self.images_fps += img_src
            elif len(mask_src):
                print(f"Skipping {dirpath} as it contains masks but no images.")

        assert len(self.images_fps) != 0, f"{dataset_path} does not contain any images."
        self.actaul_size = len(self.mask_fps)
        self.augmentation = augmentation

    def __getitem__(self, i):
        j = i % self.actaul_size
        image = cv2.imread(self.images_fps[j])
        h, w, _ = image.shape
        h, w = 256, 256

        image = cv2.resize(image, (w, h))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = image.astype('float32') / 255.0

        mask1 = cv2.imread(self.mask_fps[j], 0)
        mask1 = cv2.resize(mask1, (w, h), interpolation=cv2.INTER_NEAREST)
        mask = [(mask1 == ((i + 1) * 100)) for i in range(self.n_classes)]

        mask = np.stack(mask, axis=-1)
        mask = mask.astype('float32')

        background = 1.0 * (mask1 == 0)
        deciduous = 1.0 * (mask1 == 100)
        coniferous = 1.0 * (mask1 == 200)
        mask = np.stack([background, deciduous, coniferous], axis=-1)
        mask = 1.0 * (mask > 0.9)

        if self.augmentation and i >= self.actaul_size:
            sample = self.augmentation(image=image.astype('float32'), mask=mask.astype('float32'))
            image, mask = sample['image'], sample['mask']

        return np.asarray(image, dtype=np.float32), np.asarray(mask, dtype=np.float32)

    def __len__(self):
        return self.actaul_size * 2 if self.augmentation else self.actaul_size
```

```python
class Dataloader(tf.keras.utils.Sequence):
    def __init__(self, indexes, dataset, batch_size=1, shuffle=False):
        self.dataset = dataset
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indexes = indexes
        self.on_epoch_end()

    def __getitem__(self, i):
        start, stop = i * self.batch_size, (i + 1) * self.batch_size
        data = [self.dataset[self.indexes[j]] for j in range(start, stop)]
        return tuple(np.stack(samples, axis=0) for samples in zip(*data))

    def __len__(self):
        return len(self.indexes) // self.batch_size

    def on_epoch_end(self):
        if self.shuffle:
            self.indexes = np.random.permutation(self.indexes)

def aug():
    train_transform = [
        A.ShiftScaleRotate(scale_limit=0.1, rotate_limit=5, shift_limit=0.1, p=0.9, border_mode=0),
        A.HorizontalFlip(p=0.5),
        A.OneOf([
            A.Blur(blur_limit=3, p=1),
            A.MotionBlur(blur_limit=3, p=1),
        ], p=0.5),
        A.OneOf([
            A.HueSaturationValue(p=1),
        ], p=0.9),
        A.OneOf([
            A.CLAHE(p=1),
            A.RandomGamma(p=1),
        ], p=0.9),
    ]
    return A.Compose(train_transform)

def visualize(**images):
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
    plt.show()
```

```python
def create_model(num_classes):
    base_model = tf.keras.applications.EfficientNetB3(include_top=False, weights='imagenet', input_shape=(256, 256, 3))
    base_model.trainable = True

    inputs = tf.keras.layers.Input(shape=(256, 256, 3))
    x = base_model(inputs, training=True)

    # Upsample to original image size using Conv2DTranspose
    x = tf.keras.layers.Conv2DTranspose(512, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)  # 8x8 -> 16x16
    x = tf.keras.layers.Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)  # 16x16 -> 32x32
    x = tf.keras.layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)  # 32x32 -> 64x64
    x = tf.keras.layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)    # 64x64 -> 128x128
    x = tf.keras.layers.Conv2DTranspose(32, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)    # 128x128 -> 256x256

    # Final output layer
    outputs = tf.keras.layers.Conv2D(num_classes, (1, 1), activation='softmax')(x)  # Output shape (256, 256, num_classes)

    model = tf.keras.models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model



model = create_model(3)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=[
        tf.keras.metrics.MeanIoU(num_classes=3),
        tf.keras.metrics.CategoricalAccuracy(),
    ]
)

```

```python
import tensorflow as tf

def build_efficient_segmentation_model(input_shape=(256, 256, 3), num_classes=3):
    """
    Build a semantic segmentation model using EfficientNetB3 as encoder with corrected dimensions.
    """
    base_model = tf.keras.applications.EfficientNetB3(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )

    # Get encoder stages with correct dimensions
    stage1 = base_model.get_layer('block1b_add').output           # 128x128, 24 channels
    stage2 = base_model.get_layer('block2b_add').output           # 64x64, 32 channels
    stage3 = base_model.get_layer('block3b_add').output           # 32x32, 48 channels
    stage4 = base_model.get_layer('block5b_add').output           # 16x16, 136 channels
    stage5 = base_model.output                                    # 8x8, 1536 channels

    # Decoder path
    # Stage 5 -> 4 (8x8 -> 16x16)
    x = tf.keras.layers.Conv2DTranspose(256, 2, strides=2, padding='same')(stage5)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Project stage4 to match channels
    stage4_proj = tf.keras.layers.Conv2D(256, 1, padding='same')(stage4)
    stage4_proj = tf.keras.layers.BatchNormalization()(stage4_proj)
    stage4_proj = tf.keras.layers.ReLU()(stage4_proj)

    x = tf.keras.layers.Concatenate()([x, stage4_proj])
    x = tf.keras.layers.Conv2D(256, 3, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Stage 4 -> 3 (16x16 -> 32x32)
    x = tf.keras.layers.Conv2DTranspose(128, 2, strides=2, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    stage3_proj = tf.keras.layers.Conv2D(128, 1, padding='same')(stage3)
    stage3_proj = tf.keras.layers.BatchNormalization()(stage3_proj)
    stage3_proj = tf.keras.layers.ReLU()(stage3_proj)

    x = tf.keras.layers.Concatenate()([x, stage3_proj])
    x = tf.keras.layers.Conv2D(128, 3, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Stage 3 -> 2 (32x32 -> 64x64)
    x = tf.keras.layers.Conv2DTranspose(64, 2, strides=2, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    stage2_proj = tf.keras.layers.Conv2D(64, 1, padding='same')(stage2)
    stage2_proj = tf.keras.layers.BatchNormalization()(stage2_proj)
    stage2_proj = tf.keras.layers.ReLU()(stage2_proj)

    x = tf.keras.layers.Concatenate()([x, stage2_proj])
    x = tf.keras.layers.Conv2D(64, 3, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Stage 2 -> 1 (64x64 -> 128x128)
    x = tf.keras.layers.Conv2DTranspose(32, 2, strides=2, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    stage1_proj = tf.keras.layers.Conv2D(32, 1, padding='same')(stage1)
    stage1_proj = tf.keras.layers.BatchNormalization()(stage1_proj)
    stage1_proj = tf.keras.layers.ReLU()(stage1_proj)

    x = tf.keras.layers.Concatenate()([x, stage1_proj])
    x = tf.keras.layers.Conv2D(32, 3, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Final upsampling (128x128 -> 256x256)
    x = tf.keras.layers.Conv2DTranspose(32, 2, strides=2, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Final convolution
    outputs = tf.keras.layers.Conv2D(num_classes, 1, activation='softmax')(x)

    model = tf.keras.Model(inputs=base_model.input, outputs=outputs)
    return model

# Build and compile model
model = build_efficient_segmentation_model()
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=[
        tf.keras.metrics.MeanIoU(num_classes=3),
        tf.keras.metrics.CategoricalAccuracy()
    ]
)
```

```python
import tensorflow as tf

def build_segmentation_model(input_shape=(256, 256, 3), num_classes=3):
    """
    Build a semantic segmentation model using EfficientNetB3 as encoder with proper skip connections.

    Args:
        input_shape: Input image dimensions (height, width, channels)
        num_classes: Number of segmentation classes
    """
    # Load the pre-trained EfficientNetB3 as the encoder
    base_model = tf.keras.applications.EfficientNetB3(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )

    # Freeze the encoder layers
    base_model.trainable = False

    # Extract features from different layers for skip connections
    # Choose layers with progressively smaller spatial dimensions
    skip_connections = [
        base_model.get_layer('block2a_expand_activation').output,  # 128x128
        base_model.get_layer('block3a_expand_activation').output,  # 64x64
        base_model.get_layer('block4a_expand_activation').output,  # 32x32
        base_model.get_layer('block6a_expand_activation').output   # 16x16
    ]

    # Encoder output
    encoder_output = base_model.output  # 8x8

    # Initial number of filters for decoder
    filters = 256

    # Decoder with skip connections
    x = encoder_output

    # Upsample and combine with skip connections
    for skip_connection in reversed(skip_connections):
        # Upsampling block
        x = tf.keras.layers.Conv2DTranspose(
            filters=filters,
            kernel_size=(3, 3),
            strides=(2, 2),
            padding='same',
            use_bias=False
        )(x)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.ReLU()(x)

        # Project skip connection to the right number of channels
        skip_connection = tf.keras.layers.Conv2D(
            filters=filters,
            kernel_size=(1, 1),
            padding='same',
            use_bias=False
        )(skip_connection)
        skip_connection = tf.keras.layers.BatchNormalization()(skip_connection)

        # Concatenate
        x = tf.keras.layers.Concatenate()([x, skip_connection])

        # Refine features
        x = tf.keras.layers.Conv2D(
            filters=filters,
            kernel_size=(3, 3),
            padding='same',
            use_bias=False
        )(x)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.ReLU()(x)

        # Reduce number of filters for next iteration
        filters = filters // 2

    # Final upsampling to match input resolution
    x = tf.keras.layers.Conv2DTranspose(
        filters=64,
        kernel_size=(3, 3),
        strides=(2, 2),
        padding='same'
    )(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Final convolution
    outputs = tf.keras.layers.Conv2D(
        filters=num_classes,
        kernel_size=(1, 1),
        padding='same',
        activation='softmax'
    )(x)

    # Create model
    model = tf.keras.Model(inputs=base_model.input, outputs=outputs)

    return model

# Build and compile the model
model = build_segmentation_model()
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=[
        tf.keras.metrics.MeanIoU(num_classes=3),
        tf.keras.metrics.CategoricalAccuracy(),
    ],
)

# Print model summary
model.summary()
```
